# **1. Histogram**

Histogram is one of the common visualization techniques that is used for visualizing the distribution of a particular data. In OpenCV histograms can be used for plotting the intensity of pixel values in an image. It gives valuable information about the tone of the image which is useful in various image processing tasks like thresholding, enhancement, etc.



```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Read the image in grayscale
image = cv.imread('D:/OpenCV2/Assets/newton.jpg')

# Compute the histogram
histogram = cv.calcHist(images = [image], channels=[0], mask = None, histSize = [256], ranges=[0, 256])

# Plot the histogram
plt.plot(histogram, color='black')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.title('Grayscale Histogram')
plt.grid(True)
plt.show()
```

## **1.2 Color Histogram**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image in BGR format
image = cv2.imread('D:/OpenCV2/Assets/newton.jpg')

# Split the image into its color channels
blue_channel, green_channel, red_channel = cv2.split(image)

# Compute histograms for each color channel
blue_hist = cv2.calcHist([blue_channel], [0], None, [256], [0, 256])
green_hist = cv2.calcHist([green_channel], [0], None, [256], [0, 256])
red_hist = cv2.calcHist([red_channel], [0], None, [256], [0, 256])

# Plot the histograms
plt.plot(blue_hist, color='blue', label='Blue')
plt.plot(green_hist, color='green', label='Green')
plt.plot(red_hist, color='red', label='Red')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.title('RGB Color Histogram')
plt.legend()
plt.grid(True)
plt.show()
```
